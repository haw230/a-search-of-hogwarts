```python
with open('analysis.txt', encoding='utf8') as f:
    data = f.read()
```

```python
len(data.split())
```

```python
all_data = data.split()

from collections import defaultdict

count = defaultdict(int)


for keyword in all_data:
    s = keyword
    if len(s) < 3 or s in {'a', 'the', 'an', 'it'}:
        continue
    s = s.lower()
    count[s] += 1

items = list(count.items())

for elem in sorted(items, key=lambda x: x[1], reverse=True)[:100]:
    print(f'{elem[0]}: {elem[1]}')
```

```python
for elem in sorted(items, key=lambda x: x[1], reverse=False)[:10000]:
    print(f'{elem[0]}: {elem[1]}')
```

```python
for elem in sorted(items, key=lambda x: len(x[0]), reverse=True)[:1000]:
    print(f'{elem[0]}: {elem[1]}')
```
